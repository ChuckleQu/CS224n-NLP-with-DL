### (a)
Note that the entry of $\boldsymbol{y}$ is 1 at the position corresponding to the word vector $o$, and 0
else where. Therefore, the given equation and be transformed into:

$$-\sum_{w\in Vocab} y_w\log\left(\hat{y}_w\right) = - (1) \cdot \log\left(\hat{y}_o\right) = - \log\left(\hat{y}_o\right)$$

### (b)
$
\boldsymbol{U}=
\begin{bmatrix}
&u_{1} \\
&\vdots \\
&u_{o}\\
&\vdots \\
&u_{|Vocab|}
\end{bmatrix}_{|Vocab|\times d}
$

$\boldsymbol V=
\begin{bmatrix}
&v_{1}\\
&\vdots\\
&v_{c}\\
&\vdots\\
&v_{|Vocab|}
\end{bmatrix}_{|Vocab|\times d}
$

$
\boldsymbol y=
\begin{bmatrix}
&0\\
&\vdots\\
&1\\
&\vdots\\
&0
\end{bmatrix}_{|Vocab|\times 1}
$

$
\hat{\boldsymbol{y}} = P\left(O|C=c\right) =
\text{softmax}\left(\boldsymbol{U} \boldsymbol{v}_c^\top \right)
$

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{J}_{\text{naive-softmax}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=-\frac{\partial}{\partial\boldsymbol{v}_c}\log\frac{\exp\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)}
{\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}\\
&=-\frac{\partial}{\partial\boldsymbol{v}_c}\left[\log\exp\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)
-\log\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)\right]\\
&=-\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{u}_o^\top\boldsymbol{v}_c+\frac{\partial}{\partial\boldsymbol{v}_c}
\log\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{u}_o^\top\boldsymbol{v}_c&=\boldsymbol{u}_o\\
&=\boldsymbol{y}^\top\boldsymbol{U}
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\log\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}\frac{\partial}{\partial\boldsymbol{v}_c}
\sum\nolimits_{x\in Vocab}\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}
\sum\nolimits_{x\in Vocab}\frac{\partial}{\partial\boldsymbol{v}_c}\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}
\sum\nolimits_{x\in Vocab}\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)\boldsymbol{u}_x\\
&=\sum\nolimits_{x\in Vocab}\frac{\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)}{\sum\nolimits_{w\in Vocab}
\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}\boldsymbol{u}_x\\
&=\sum\nolimits_{x\in Vocab}P\left(O=x|C=c\right)\boldsymbol{u}_x\\
&=\sum\nolimits_{x\in Vocab}\hat{\boldsymbol{y}}_x\boldsymbol{u}_x\\
&=\hat{\boldsymbol{y}}^\top\boldsymbol{U}
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{J}_{\text{naive-softmax}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=-\boldsymbol{y}^\top\boldsymbol{U}+\hat{\boldsymbol{y}}^\top\boldsymbol{U}\\
&=\left(\hat{\boldsymbol{y}}^\top- \boldsymbol{y}^\top\right)\boldsymbol{U}
\end{aligned}

\begin{aligned}

&=\frac{1}{\sum\nolimits_{x\in Vocab}\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)}\frac{\partial}{\partial\boldsymbol{u}_w}
\sum\nolimits_{z\in Vocab}\exp\left(\boldsymbol{u}_z^\top\boldsymbol{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{x\in Vocab}\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)}
\sum\nolimits_{z\in Vocab}\frac{\partial}{\partial\boldsymbol{u}_w}\exp\left(\boldsymbol{u}_z^\top\boldsymbol{v}_c\right)\\

\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)\boldsymbol{v}_c\\
&=\frac{\exp\left(\boldsymbol{u}_w^\top\boldsymbol{v}_c\right)}{\sum\nolimits_{x\in Vocab}
\exp\left(\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)}\boldsymbol{v}_c\\
&=P\left(O=w|C=c\right)\boldsymbol{v}_c\\
&=\hat{\boldsymbol{y}}_w\boldsymbol{v}_c\\
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{u}_w}\boldsymbol{J}_{\text{naive-softmax}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)

\begin{cases}

\hat{\boldsymbol{y}}_w\boldsymbol{v}_c ,& \text{when } w \neq o
\end{cases}
=
\begin{cases}
\left(\hat{\boldsymbol{y}}_w-1\right)\boldsymbol{v}_c ,& \text{when } w=o\\
\hat{\boldsymbol{y}}_w\boldsymbol{v}_c ,& \text{when } w \neq o
\end{cases}\\
\end{aligned}

\begin{aligned}
\frac{\mathrm{d}}{\mathrm{d}x}\sigma (x)
&=\frac{\mathrm{d}}{\mathrm{d}x}\left(1+e^{-x}\right)^{-1}\\
&=-\left(1+e^{-x}\right)^{-2}\left(-e^{-x}\right)\\
&=\frac{e^{-x}}{\left(1+e^{-x}\right)^{2}}\\
&=\frac{1+e^{-x}}{\left(1+e^{-x}\right)^{2}}-\frac{1}{\left(1+e^{-x}\right)^{2}}\\
&=\frac{1}{1+e^{-x}}-\frac{1}{\left(1+e^{-x}\right)^{2}}\\
&=\frac{1}{1+e^{-x}}\left(1-\frac{1}{1+e^{-x}}\right)\\
&=\sigma (x)\left(1-\sigma (x)\right)
\end{aligned}

### 1. Partial derivatives of $\boldsymbol{J}_{\text{neg-sample}}$ with respect to $\boldsymbol{v}_c$

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{J}_{\text{neg-sample}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=\frac{\partial}{\partial\boldsymbol{v}_c}\left[-\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
-\sum_{k=1}^K\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)\right]\\
&=-\frac{\partial}{\partial\boldsymbol{v}_c}\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
-\frac{\partial}{\partial\boldsymbol{v}_c}\sum_{k=1}^K\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
&=\frac{1}{\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)}\left[\frac{\partial}{\partial\boldsymbol{v}_c}\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right]\\
&=\frac{1}{\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)}\left[\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\left(1-\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\right)\boldsymbol{u}_o\right]\\
&=\left( 1-\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\right)\boldsymbol{u}_o
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\sum_{k=1}^K\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)
&=\sum_{k=1}^K\frac{\partial}{\partial\boldsymbol{v}_c}\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)\\
&=\sum_{k=1}^K\left( 1-\sigma (-\boldsymbol{u}_k^\top\boldsymbol{v}_c)\right)\left(-\boldsymbol{u}_k\right)
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{v}_c}\boldsymbol{J}_{\text{neg-sample}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=-\left( 1-\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\right)\boldsymbol{u}_o-
\sum_{k=1}^K\left( 1-\sigma (-\boldsymbol{u}_k^\top\boldsymbol{v}_c)\right)\left(-\boldsymbol{u}_k\right)
\end{aligned}

### 2. Partial derivatives of $\boldsymbol{J}_{\text{neg-sample}}$ with respect to $\boldsymbol{u}_o$

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{u}_o}\boldsymbol{J}_{\text{neg-sample}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=\frac{\partial}{\partial\boldsymbol{u}_o}\left[-\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
-\sum_{k=1}^K\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)\right]\\
&=-\frac{\partial}{\partial\boldsymbol{u}_o}\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
-\frac{\partial}{\partial\boldsymbol{u}_o}\sum_{k=1}^K\log\left(\sigma\left(-\boldsymbol{u}_k^\top\boldsymbol{v}_c\right)\right)\\
&=-\frac{1}{\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)}\left[\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\left(1-\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\right)\boldsymbol{v}_c\right]-0\\
&=-\left( 1-\sigma (\boldsymbol{u}_o^\top\boldsymbol{v}_c)\right)\boldsymbol{v}_c
\end{aligned}

### 3. Partial derivatives of $\boldsymbol{J}_{\text{neg-sample}}$ with respect to $\boldsymbol{u}_k$

\begin{aligned}
\frac{\partial}{\partial\boldsymbol{u}_k}\boldsymbol{J}_{\text{neg-sample}}\left(\boldsymbol{v}_c, o, \boldsymbol{U}\right)
&=\frac{\partial}{\partial\boldsymbol{u}_k}\left[-\log\left(\sigma\left(\boldsymbol{u}_o^\top\boldsymbol{v}_c\right)\right)
-\sum_{x=1}^K\log\left(\sigma\left(-\boldsymbol{u}_x^\top\boldsymbol{v}_c\right)\right)\right]\\
