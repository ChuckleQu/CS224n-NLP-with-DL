## Written: Understanding
### (a)
Note that the entry of $\pmb{y}$ is 1 at the position corresponding to the word vector $o$, and 0
else where. Therefore, the given equation and be transformed into:

$$-\sum_{w\in Vocab} y_w\log\left(\hat{y}_w\right) = - (1) \cdot \log\left(\hat{y}_o\right) = - \log\left(\hat{y}_o\right)$$

### (b)
$
\pmb{U}=
\begin{bmatrix}
&u_{1} \\
&\vdots \\
&u_{o}\\
&\vdots \\
&u_{|Vocab|}
\end{bmatrix}_{|Vocab|\times d}
$

$\pmb V=
\begin{bmatrix}
&v_{1}\\
&\vdots\\
&v_{c}\\
&\vdots\\
&v_{|Vocab|}
\end{bmatrix}_{|Vocab|\times d}
$

$
\pmb y=
\begin{bmatrix}
&0\\
&\vdots\\
&1\\
&\vdots\\
&0
\end{bmatrix}_{|Vocab|\times 1}
$

$
\hat{\pmb{y}} = P\left(O|C=c\right) =
\text{softmax}\left(\pmb{U} \pmb{v}_c^\top \right)
$

\begin{aligned}
\frac{\partial}{\partial\pmb{v}_c}\pmb{J}_{\text{naive-softmax}}\left(\pmb{v}_c, o, \pmb{U}\right)
&=-\frac{\partial}{\partial\pmb{v}_c}\log\frac{\exp\left(\pmb{u}_o^\top\pmb{v}_c\right)}
{\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}\\
&=-\frac{\partial}{\partial\pmb{v}_c}\left[\log\exp\left(\pmb{u}_o^\top\pmb{v}_c\right)
-\log\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)\right]\\
&=-\frac{\partial}{\partial\pmb{v}_c}\pmb{u}_o^\top\pmb{v}_c+\frac{\partial}{\partial\pmb{v}_c}
\log\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{v}_c}\pmb{u}_o^\top\pmb{v}_c&=\pmb{u}_o\\
&=\pmb{y}^\top\pmb{U}
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{v}_c}\log\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}\frac{\partial}{\partial\pmb{v}_c}
\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}
\sum\nolimits_{x\in Vocab}\frac{\partial}{\partial\pmb{v}_c}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{w\in Vocab}\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}
\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)\pmb{u}_x\\
&=\sum\nolimits_{x\in Vocab}\frac{\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}{\sum\nolimits_{w\in Vocab}
\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}\pmb{u}_x\\
&=\sum\nolimits_{x\in Vocab}P\left(O=x|C=c\right)\pmb{u}_x\\
&=\sum\nolimits_{x\in Vocab}\hat{\pmb{y}}_x\pmb{u}_x\\
&=\hat{\pmb{y}}^\top\pmb{U}
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{v}_c}\pmb{J}_{\text{naive-softmax}}\left(\pmb{v}_c, o, \pmb{U}\right)
&=-\pmb{y}^\top\pmb{U}+\hat{\pmb{y}}^\top\pmb{U}\\
&=\left(\hat{\pmb{y}}^\top- \pmb{y}^\top\right)\pmb{U}
\end{aligned}

### (c)

\begin{aligned}
\frac{\partial}{\partial\pmb{u}_w}\pmb{J}_{\text{naive-softmax}}\left(\pmb{v}_c, o, \pmb{U}\right)
&=-\frac{\partial}{\partial\pmb{u}_w}\log\frac{\exp\left(\pmb{u}_o^\top\pmb{v}_c\right)}
{\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}\\
&=-\frac{\partial}{\partial\pmb{u}_w}\left[\log\exp\left(\pmb{u}_o^\top\pmb{v}_c\right)
-\log\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)\right]\\
&=-\frac{\partial}{\pmb{u}_w}\pmb{u}_o^\top\pmb{v}_c+\frac{\partial}{\partial\pmb{u}_w}
\log\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{u}_w}\pmb{u}_o^\top\pmb{v}_c=
\begin{cases}
\pmb{v}_c ,& \text{when } w=o\\
0 ,& \text{when } w \neq o
\end{cases}
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{u}_w}\log\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)
&=\frac{1}{\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}\frac{\partial}{\partial\pmb{u}_w}
\sum\nolimits_{z\in Vocab}\exp\left(\pmb{u}_z^\top\pmb{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}
\sum\nolimits_{z\in Vocab}\frac{\partial}{\partial\pmb{u}_w}\exp\left(\pmb{u}_z^\top\pmb{v}_c\right)\\
&=\frac{1}{\sum\nolimits_{x\in Vocab}\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}
\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)\pmb{v}_c\\
&=\frac{\exp\left(\pmb{u}_w^\top\pmb{v}_c\right)}{\sum\nolimits_{x\in Vocab}
\exp\left(\pmb{u}_x^\top\pmb{v}_c\right)}\pmb{v}_c\\
&=P\left(O=w|C=c\right)\pmb{v}_c\\
&=\hat{\pmb{y}}_w\pmb{v}_c\\
\end{aligned}

\begin{aligned}
\frac{\partial}{\partial\pmb{v}_c}\pmb{J}_{\text{naive-softmax}}\left(\pmb{v}_c, o, \pmb{U}\right)
=
\begin{cases}
-\pmb{v}_c+\hat{\pmb{y}}_w\pmb{v}_c ,& \text{when } w=o\\
\hat{\pmb{y}}_w\pmb{v}_c ,& \text{when } w \neq o
\end{cases}
=
\begin{cases}
\left(\hat{\pmb{y}}_w-1\right)\pmb{v}_c ,& \text{when } w=o\\
\hat{\pmb{y}}_w\pmb{v}_c ,& \text{when } w \neq o
\end{cases}\\
\end{aligned}
